```python
#importing basic liberaries
import numpy as np
import pandas as pd
```

```python
#creating headers for the data
headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv("dataset_1.data",names = headers_data)
df
```

```python
#checking for any missing values or some problem with the dataset
df.info()
```

```python
df.describe(include = "all")
```

```python
#replacing '?' with np.nan so that we can change its data type
df.replace("?",np.nan, inplace = True)
```

```python
df.info()
```

```python
#changing the datatype of the object columns
avg_nor_loss = df["normalized-losses"].astype("float").mean()
df["normalized-losses"].replace(np.nan,avg_nor_loss,inplace = True)
```

```python
df.head(5)
```

```python
avg_bore = df["bore"].astype("float").mean()
df["bore"].replace(np.nan,avg_bore,inplace = True)
```

```python
avg_stroke = df["stroke"].astype("float").mean()
df["stroke"].replace(np.nan,avg_stroke,inplace = True)
```

```python
avg_horsepower = df["horsepower"].astype("float").mean()
df["horsepower"].replace(np.nan,avg_horsepower,inplace = True)
```

```python
avg_peak_rpm=df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_peak_rpm, inplace=True)
```

```python
df["num-of-doors"].replace(np.nan,"four",inplace = True)
```

```python
#droping some rows where price is not given because it will not be helpful for us
df.dropna(subset = ["price"],inplace = True,axis = 0)
df.reset_index(drop = True, inplace = True)
```

```python
df.info()
```

```python
df.to_csv("polissed_df.csv")
```

```python
#binning the data
import matplotlib.pyplot as plt
plt.hist(df_1["horsepower"])
plt.xlabel("horsepower")
plt.ylabel("No of cars")
```

```python
np.linspace(1,10,4) # this concept can be used in binning
```

```python
#creating bins for the horsepower column
bins = np.linspace(min(df_1["horsepower"]),max(df_1["horsepower"]),4)
bins
```

```python
#sorting data based on the horsepower of the dataset
grp_names = ["low","medium","high"]
df_1["horsepower_binned"] = pd.cut(df_1["horsepower"],bins,labels = grp_names,include_lowest = True)
df_1[["horsepower","horsepower_binned"]].tail(30)
```

```python
#visualizing the data using matplotlib
plt.bar(grp_names,df_1["horsepower_binned"].value_counts())
plt.ylabel("Number of cars")
```
